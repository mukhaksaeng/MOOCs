## 1. Logistic Regression

In this part of the exercise, we will build a logistic regression model to predict whether a student gets admitted into a university. Suppose that you are the administrator of a university department and you want to determine each applicant’s chance of admission based on their results on two exams. You have historical data from previous applicants
that you can use as a training set for logistic regression (stored in `ex2data1.txt`). For each training example, you have the applicant’s scores on two exams and the admissions decision.

In line with this, we will build a classification model that estimates an applicant’s probability of admission based the scores from those two exams.

### 1.1 Visualizing the Data

Before starting to implement any learning algorithm, it is always good to visualize the data if possible.

```python
# Read comma-separated data, no headers
data = pd.read_csv("ex2data1.txt", sep = ",", header = None)

# Separate into X and y values
X = data[[0, 1]]
y = data[2][:, np.newaxis]
```

```python
# Plot the data
pos = [i for i in range(len(y)) if y[i] == 1]
neg = [i for i in range(len(y)) if y[i] == 0]

plt.scatter(X.loc[pos, 0], X.loc[pos, 1], marker = "+", s = 30, linewidths = 2, color = "black", label = "Admitted")
plt.scatter(X.loc[neg, 0], X.loc[neg, 1], marker = "o", s = 30, color = "yellow", linewidths = 0.5, edgecolor = "black", label = "Not Admitted")
plt.xlabel("Exam 1 score")
plt.ylabel("Exam 2 score")
plt.legend()
```

### 1.2 Implementation

#### 1.2.1 Sigmoid Function

Recall that the logistic regression hypothesis is defined as:
    
$\begin{equation}
h_{\theta}(x) = g(\theta^{T}x)
\end{equation}$

where the function $g$ is the sigmoid function, which is defined as:

$\begin{equation}
g(z) = \frac{1}{1 + e^{-z}}
\end{equation}$

We implement this function in `sigmoid` so that it can be called on by the succeeding functions.

```python
# sigmoid computes the sigmoid of each value of z (z can be a matrix, vector or scalar).
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

#### 1.2.2 Cost Function and Gradient

Recall that the cost function in logistic regression hypothesis is:
    
$\begin{equation}
J(\theta) = \frac{1}{m}\sum_{i = 1}^{m}[-y^{(i)}\log(h_{\theta}(x^{(i)})-(1 - y^{(i)})\log(1 - h_{\theta}(x^{(i)})]
\end{equation}$

and the gradient of the cost is a vector of the same length as $\theta$ where the $j^{th}$ element (for $j = 0, 1, ..., n)$ is defined as follows:

$\begin{equation}
\frac{\partial J(\theta)}{\partial\theta_{j}} = \frac{1}{m}\sum_{i = 1}^{m}(h_{\theta}(x^{(i)})-  y^{(i)})x_{j}^{(i)}
\end{equation}$

Note that while this gradient looks identical to the linear regression gradient, the formula is actually different because linear and logistic regression have different definitions of $h_{\theta}(x)$.

We implement the above formulas in the function `costFunction` below.

```python
# costFunction computes the cost of using theta as the parameter for logistic regression and the gradient of the cost
# w.r.t. to the parameters.
def costFunction(theta, X, y):
    # Initialize values
    m = len(y)
    J = 0
    grad = np.zeros(theta.shape[0])
    
    # Compute cost
    prediction = sigmoid(X @ theta)
    J = -(1 / m) * (y.T @ np.log(prediction) + (1 - y).T @ np.log(1 - prediction))
    
    # Compute gradient
    error = prediction - y
    grad = (1 / m) * np.sum(error.T * X.T, axis = 1)[:, np.newaxis]

    return np.asscalar(J), grad
```

```python
# Test

# Initialize values and add column of 1's to X
m, n = X.shape
X = np.c_[np.ones((m, 1)), X]
```

```python
# Initialize fitting parameters
initial_theta = np.zeros((n + 1, 1))

# Compute and display initial cost and gradient
cost, grad = costFunction(initial_theta, X, y)
print("Cost at initial theta (zeros): ", cost, "\n")
print("Expected cost (approx): 0.693\n")
print("Gradient at initial theta (zeros): ")
print(grad[0, 0], "\n", grad[1, 0], "\n", grad[2, 0], "\n")
print("Expected gradients (approx):\n -0.1000\n -12.0092\n -11.2628\n")

# Compute and display cost and gradient with non-zero theta
test_theta = np.array([-24, 0.2, 0.2]).reshape(-1, 1)
cost, grad = costFunction(test_theta, X, y);

print("Cost at initial theta (zeros): ", cost, "\n")
print("Expected cost (approx): 0.218\n")
print("Gradient at initial theta (zeros): ")
print(grad[0, 0], "\n", grad[1, 0], "\n", grad[2, 0], "\n")
print("Expected gradients (approx):\n 0.043\n 2.566\n 2.647\n\n")
```

#### 1.2.3 Learning Parameters using `minimize`

In the previous assignment, we found the optimal parameters of a linear regression model by implementing gradent descent. This time, instead of taking gradient descent steps, we will use a built-in function. MATLAB/Octave has the optimization solver `fminunc` that finds the minimum of an unconstrained function. Its Python equivalent is the `minimize` command in the `scipy.optimize` package. 

For logistic regression, you want to optimize the cost function $J(\theta)$ with parameters $\theta$. The `minimize` function in this case takes the following inputs:
- The dataset (X, y) (argument `args`)
- The optimization method (argument `method`)
- The initial values of the parameters we are trying to optimize (argument `x0`)
- A function that, when given the training set and a particular $\theta$, computes the logistic regression cost (argument `fun`)
- A function that, when given the training set and a particular $\theta$, computes the gradient with respect to $\theta$ (argument `jac`)

In this case, we will have to separate the `costFunction` function from earlier into two -- one of which will output the cost (`costFunction` below) and the other outputting the gradient (`gradient` below). Since the argument `x0` can only take in 1D arrays, we modify the functions so that `theta` and `grad` are 1D arrays. 

Furthermore, we will use BFGS as our optimization method.

We implement the optimization below and store the optimum cost and $\theta$.

```python
# Initialize parameters
initial_theta = np.zeros(n + 1)

def costFunction(theta, X, y):
    # Initialize values
    m = len(y)
    J = 0
    
    # Compute cost
    prediction = sigmoid(X @ theta[:, np.newaxis])
    J = -(1 / m) * (y.T @ np.log(prediction) + (1 - y).T @ np.log(1 - prediction))
    
    return np.asscalar(J)

def gradient(theta, X, y):
    # Initialize values
    grad = np.zeros(theta.shape[0])
    
    # Compute gradient
    prediction = sigmoid(X @ theta[:, np.newaxis])
    error = prediction - y
    grad = (1 / m) * np.sum(error.T * X.T, axis = 1)

    return grad

result = op.minimize(fun = costFunction, x0 = initial_theta, args = (X, y), method = "BFGS", jac = gradient)
opt_cost = result.fun
opt_theta = result.x
print("cost = ", opt_cost)
print("theta = ", opt_theta)
```

The final $\theta$ value that we obtained after optimization can now be used to plot the decision boundary on the
training data.

Note that for a linear decision boundary, the hypothesis is given by 

$\begin{equation}
h_{\theta}(x) = \theta_{0} + \theta_{1}x_{1} + \theta_{2}x_{2}
\end{equation}$

At the decision boundary, $h_{\theta}(x) = 0$. We then cast the resulting equation into the equation of a line, with $x_{1}$ in the x-axis and $x_{2}$ in the y-axis. As such, it becomes

$\begin{equation}
x_{2} = -\frac{1}{\theta_{2}}(\theta_{0} + \theta_{1}x_{1})
\end{equation}$

We can now plot the decision boundary and the data, as is done below.

```python
# Plot the data
X_data = data[[0, 1]]

pos = [i for i in range(len(y)) if y[i] == 1]
neg = [i for i in range(len(y)) if y[i] == 0]

plt.scatter(X_data.loc[pos, 0], X_data.loc[pos, 1], marker = "+", s = 30, linewidths = 2, color = "black", label = "Admitted")
plt.scatter(X_data.loc[neg, 0], X_data.loc[neg, 1], marker = "o", s = 30, color = "yellow", linewidths = 0.5, edgecolor = "black", label = "Not Admitted")

# Plot the decision boundary
# Only need 2 points to define a line, so choose two endpoints
plot_x = np.array([np.min(X_data[0]) - 2,  np.max(X_data[0]) + 2])

# Calculate the decision boundary line
plot_y = (-1 / opt_theta[2]) * (opt_theta[1] * plot_x + opt_theta[0])

# Plot the decisioon boundary
plt.plot(plot_x, plot_y, label = "Decision Boundary")
    
# Annotate graph
plt.xlabel("Exam 1 score")
plt.ylabel("Exam 2 score")
plt.legend()    
```

#### 1.2.4 Evaluating logistic regression

After learning the parameters, you can use the model to predict whether a particular student will be admitted. For a  student with an Exam 1 score of 45 and an Exam 2 score of 85, you should expect to see an admission probability of 0.776, calculated as follows:

```python
prob = sigmoid(np.array([1, 45, 85]) @ opt_theta)
print("For a student with scores 45 and 85, we predict an admission probability of ", prob, "\n")
print("Expected value: 0.776")
```

Another way to evaluate the quality of the parameters we have found is to see how well the learned model predicts on our training set, which is the training accuracy of the classifier. To help with this, we need to implement a function which predicts the outcome (either "0" or "1") based on the resulting parameters $\theta$. Such is the case with the `predict` function below.

```python
# predict predicts whether the label is 0 or 1 using learned logistic regression parameters theta
def predict(X, theta):
    prediction = sigmoid(X @ theta)
    p = (prediction >= 0.5).astype(int)[:, np.newaxis]
    return p
```

```python
# Compute accuracy
p = predict(X, opt_theta)
acc = np.mean(np.equal(p, y).astype(int)) * 100

print("Train Accuracy: ", acc, "\n")
print("Expected accuracy (approx): 89.0\n")
```

## 2. Regularized Logistic Regression

In this part of the exercise, we will implement regularized logistic regression to predict whether microchips from a fabrication plant passes quality assurance (QA). During QA, each microchip goes through various tests to ensure it is functioning correctly.

Suppose you are the product manager of the factory and you have the test results for some microchips on two different tests. From these two tests, you would like to determine whether the microchips should be accepted or rejected. To help you make the decision, you have a dataset of test results on past microchips, from which you can build a logistic regression model (stored in `ex2data2.txt`).

### 2.1 Visualizing the data

Again, we visualize the data before we implement logistic regression.

```python
# Read comma-separated data, no headers
data = pd.read_csv("ex2data2.txt", sep = ",", header = None)

# Separate into X and y values
X = data[[0, 1]]
y = data[2][:, np.newaxis]
```

```python
# Plot the data
pos = [i for i in range(len(y)) if y[i] == 1]
neg = [i for i in range(len(y)) if y[i] == 0]

plt.scatter(X.loc[pos, 0], X.loc[pos, 1], marker = "+", s = 30, linewidths = 2, color = "black", label = "y = 1")
plt.scatter(X.loc[neg, 0], X.loc[neg, 1], marker = "o", s = 30, color = "yellow", linewidths = 0.5, edgecolor = "black", label = "y = 0")
plt.xlabel("Microchip Test 1")
plt.ylabel("Microchip Test 2")
plt.legend()
```

The figure above shows that our dataset cannot be separated into positive and negative examples by a straight-line through the plot. Therefore, a straightforward application of logistic regression will not perform well on this dataset since logistic regression will only be able to find a linear decision boundary.

### 2.2 Feature mapping

One way to fit the data better is to create more features from each data point. As in the function `mapFeature` below, we will map the features into all polynomial terms of $x_{1}$ and $x_{2}$ up to the sixth power. That is, 

$\begin{equation}
mapFeature(x) = 
\begin{pmatrix}
1 && x_{1} && x_{2} && x_{1}^{2} && x_{1}x_{2} && x_{2}^{2} && x_{1}^{3} && \cdots && x_{1}x_{2}^{5} && x_{2}^{6}
\end{pmatrix}
\end{equation}$

As a result of this mapping, our vector of two features (the scores on two QA tests) has been transformed into a 28-dimensional vector (28 because $\sum_{n = 0}^{6} {n + m - 1 \choose m - 1}$ where $m = 2$). A logistic regression classifier trained on this higher-dimension feature vector will have a more complex decision boundary and will appear nonlinear when drawn in our 2-dimensional plot.

While the feature mapping allows us to build a more expressive classifier, it also more susceptible to overfitting. As a result, we need to implement regularized logistic regression to combat the overfitting problem.

```python
# mapFeature maps the two input features to quadratic features
def mapFeature(X1, X2):
    # Initialize values
    degree = 6
    out = np.ones((X1.shape[0], 1))
    
    # Create quadratic feature matrix
    for i in range(1, degree + 1):
        for j in range(i + 1):
            out = np.c_[out, (X1 ** (i - j)) * (X2 ** j)]
            
    return out
```

```python
# Add polynomial features to X
X = np.array(X)
X_quad = mapFeature(X[:, 0], X[:, 1])
```

### 2.3 Cost function and gradient

Now you will implement code to compute the cost function and gradient for regularized logistic regression.

Recall that the regularized cost function in logistic regression is

$\begin{equation}
J(\theta) = \frac{1}{m}\sum_{i = 1}^{m}[-y^{(i)}\log(h_{\theta}(x^{(i)})-(1 - y^{(i)})\log(1 - h_{\theta}(x^{(i)})] + \frac{\lambda}{2m} \sum_{j = 1}^{n}\theta_{j}^{2}
\end{equation}$

Meanwhile, the gradient of the cost function is a vector where the j th element is defined as follows:

For $j = 0$,

$\begin{equation}
\frac{\partial J(\theta)}{\partial\theta_{0}} = \frac{1}{m}\sum_{i = 1}^{m}(h_{\theta}(x^{(i)})-  y^{(i)})x_{0}^{(i)}
\end{equation}$

For $j \geq 1$,

$\begin{equation}
\frac{\partial J(\theta)}{\partial\theta_{j}} = \left( \frac{1}{m}\sum_{i = 1}^{m}(h_{\theta}(x^{(i)})-  y^{(i)})x_{j}^{(i)} \right) + \frac{\lambda}{m}\theta_{j}
\end{equation}$

These are implemented in the `costFunctionReg` function below.

```python
# costFunctionReg computes the cost of using theta as the parameter for regularized logistic regression and the gradient
# of the cost w.r.t. to the parameters. 
def costFunctionReg(theta, X, y, lmb):
    # Initialize values
    m = len(y)
    J = 0
    grad = np.zeros((theta.shape[0], 1))

    # Compute for cost
    prediction = sigmoid(X @ theta)
    J = -(1 / m) * (y.T @ np.log(prediction) + (1 - y).T @ np.log(1 - prediction)) + (lmb / (2 * m)) * (theta[1:].T @ theta[1:])

    # Compute for gradient
    error = prediction - y
    grad[0] = (1 / m) * np.sum(error * X[:, 0][:, np.newaxis])
    grad[1:] = (1 / m) * np.sum(error.T * X[:, 1:].T, axis = 1)[:, np.newaxis] + (lmb / m) * theta[1:] 
    
    return np.asscalar(J), grad
```

```python
# Tests
# Initialize fitting parameters
initial_theta = np.zeros((X_quad.shape[1], 1))

# Set regularization parameter lambda to 1
lmb = 1

# Compute and display initial cost and gradient for regularized logistic regression
cost, grad = costFunctionReg(initial_theta, X_quad, y, lmb)

print("Cost at initial theta (zeros): ", cost, "\n")
print("Expected cost (approx): 0.693\n")
print("Gradient at initial theta (zeros) - first five values only: \n")
print(grad[0, 0], "\n", grad[1, 0], "\n", grad[2, 0], "\n", grad[3, 0], "\n", grad[4, 0], "\n")
print("Expected gradients (approx) - first five values only:\n")
print("0.0085\n 0.0188\n 0.0001\n 0.0503\n 0.0115\n")

# Compute and display cost and gradient with all-ones theta and lambda = 10
test_theta = np.ones((X_quad.shape[1], 1))
cost, grad = costFunctionReg(test_theta, X_quad, y, 10)

print("Cost at test theta (zeros): ", cost, "\n")
print("Expected cost (approx): 3.16\n")
print("Gradient at test theta (zeros) - first five values only: \n")
print(grad[0, 0], "\n", grad[1, 0], "\n", grad[2, 0], "\n", grad[3, 0], "\n", grad[4, 0], "\n")
print("Expected gradients (approx) - first five values only:\n")
print("0.3460\n 0.1614\n 0.1948\n 0.2269\n 0.0922\n")
```

#### 2.3.1 Learning Parameters using `minimize`

As in the logistic regression exercise, we will again use the `minimize` function to learn the optimal parameters $\theta$.

Again, we will have to separate the `costFunctionReg` function from earlier into two -- one of which will output the cost (`costFunction` below) and the other outputting the gradient (`gradientReg` below). Furthermore, we will use L-BFGS-B as our optimization method. Furthermore, since `x0` must be a 1D array, we modify the functions accordingly to account for this, so that `theta` and `grad` are all 1-dimensional.

We implement the optimization below and store the optimum cost and $\theta$.

```python
import scipy.optimize as op

# Initialize parameters
initial_theta = np.zeros(X_quad.shape[1])
lmb = 1

def costFunctionReg(theta, X, y, lmb):
    # Initialize values
    m = len(y)
    J = 0

    # Compute for cost
    prediction = sigmoid(X @ theta[:, np.newaxis])
    J = -(1 / m) * (y.T @ np.log(prediction) + (1 - y).T @ np.log(1 - prediction)) + (lmb / (2 * m)) * (theta[1:].T @ theta[1:])

    return np.asscalar(J)

def gradientReg(theta, X, y, lmb):
    # Initialize values
    m = len(y)
    grad = np.zeros(theta.shape[0])

    # Compute for gradient
    prediction = sigmoid(X @ theta[:, np.newaxis])
    error = prediction - y
    grad[0] = (1 / m) * np.sum(error * X[:, 0][:, np.newaxis])
    grad[1:] = (1 / m) * np.sum(error.T * X[:, 1:].T, axis = 1) + (lmb / m) * theta[1:]
    
    return grad

result = op.minimize(fun = costFunctionReg, x0 = initial_theta, args = (X_quad, y, lmb), method = "L-BFGS-B", jac = gradientReg)
opt_cost = result.fun
opt_theta = result.x
print("cost = ", opt_cost)
print("theta = ", opt_theta)
```

The final $\theta$ value that we obtained after optimization can now be used to plot the decision boundary on the
training data.

Note that for a linear decision boundary, the hypothesis is given by 

$\begin{equation}
h_{\theta}(x) = \theta_{0} + \theta_{1}x_{1} + \theta_{2}x_{2}
\end{equation}$

At the decision boundary, $h_{\theta}(x) = 0$. We then cast the resulting equation into the equation of a line, with $x_{1}$ in the x-axis and $x_{2}$ in the y-axis. As such, it becomes

$\begin{equation}
x_{2} = -\frac{1}{\theta_{2}}(\theta_{0} + \theta_{1}x_{1})
\end{equation}$

We can now plot the decision boundary and the data, as is done below.

```python
# Plot the data
X_data = data[[0, 1]]

pos = [i for i in range(len(y)) if y[i] == 1]
neg = [i for i in range(len(y)) if y[i] == 0]

plt.scatter(X_data.loc[pos, 0], X_data.loc[pos, 1], marker = "+", s = 30, linewidths = 2, color = "black", label = "y = 1")
plt.scatter(X_data.loc[neg, 0], X_data.loc[neg, 1], marker = "o", s = 30, color = "yellow", linewidths = 0.5, edgecolor = "black", label = "y = 0")

# Plot the decision boundary
# Specify the grid range
u = np.linspace(-1, 1.5, 50)
v = np.linspace(-1, 1.5, 50)

# Evaluate z = theta * x over the grid
z = np.zeros((len(u), len(v)))
    
for i in range(len(u)):
    for j in range(len(v)):
        z[i, j] = mapFeature(np.array(u[i]).reshape(1, -1), np.array(v[j]).reshape(1, -1)) @ opt_theta

# Transpose z        
z = z.T

# Plot z = 0
plt.contour(u, v, z, levels = [0])
    
# Annotate graph
plt.xlabel("Microchip Test 1")
plt.ylabel("Microchip Test 2")
plt.legend()    
```

We will now try out different regularization parameters for the dataset to understand how regularization prevents overfitting.

```python
for lmb in [0, 1, 10, 30, 90, 100]:
    result = op.minimize(fun = costFunctionReg, x0 = initial_theta, args = (X_quad, y, lmb), method = "L-BFGS-B", jac = gradientReg)
    opt_theta = result.x
    
    # Plot the data
    X_data = data[[0, 1]]

    pos = [i for i in range(len(y)) if y[i] == 1]
    neg = [i for i in range(len(y)) if y[i] == 0]
    
    plt.figure()
    plt.scatter(X_data.loc[pos, 0], X_data.loc[pos, 1], marker = "+", s = 30, linewidths = 2, color = "black", label = "y = 1")
    plt.scatter(X_data.loc[neg, 0], X_data.loc[neg, 1], marker = "o", s = 30, color = "yellow", linewidths = 0.5, edgecolor = "black", label = "y = 0")
    plt.xlabel("Microchip Test 1")
    plt.ylabel("Microchip Test 2")
    plt.legend()

    # Plot the decision boundary
    # Specify the grid range
    u = np.linspace(-1, 1.5, 50)
    v = np.linspace(-1, 1.5, 50)

    # Evaluate z = theta * x over the grid
    z = np.zeros((len(u), len(v)))

    for i in range(len(u)):
        for j in range(len(v)):
            z[i, j] = mapFeature(np.array(u[i]).reshape(1, -1), np.array(v[j]).reshape(1, -1)) @ opt_theta

    # Transpose z        
    z = z.T

    # Plot z = 0
    plt.contour(u, v, z, levels = [0])
    plt.title("lambda = " + str(lmb))
```

With a small $\lambda$, we find that the classifier gets almost every training example correct, but draws a very complicated boundary, thus overfitting the data.

With a larger $\lambda$, we see a plot that shows an simpler decision boundary which still separates the positives and negatives fairly well. However, if $\lambda$ is set to too high a value, we do not seem to get a good fit and the decision boundary will not follow the data so well, thus underfitting the data.
