## 1. K-means Clustering

In this section, we will implement the K-means algorithm and use it for image compression. We will begin with an example 2D dataset to get a feel of how the K-means algorithm works.  After that, we will use the K-means algorithm for image compression by reducing the number of colors that occur in an image to only those that are most common in that image. 

### 1.1 Implementing K-means

First, we load the example dataset.

```python
# Read .mat file
data = io.loadmat('ex7data2.mat')
X = data["X"]
```

The K-means algorithm is a method to automatically cluster similar data examples together. Concretely, you are given a training set $\{x^{(1)}, \cdots, x^{(m)}\}$ (where $x^{(i)}$ is an n-dimensional vector), and want to group the data into a few cohesive “clusters”. The intuition behind K-means is an iterative procedure that starts by guessing the initial centroids, and then refines this guess by repeatedly assigning examples to their closest centroids and then recomputing the centroids based on the assignments.

The algorithm is as follows:


The inner-loop of the algorithm repeatedly carries out two steps: 
- Assigning each training example $x^{(i)}$ to its closest centroid
- Recomputing the mean of each centroid using the points assigned to it. The K-means algorithm will always converge to some final set of means for the centroids.

Note that the converged solution may not always be ideal and depends on the initial setting of the centroids. Therefore, in practice, the K-means algorithm is usually run a few times with different random initializations. One way to choose between these different solutions from different random initializations is to choose the one with the lowest cost function value (distortion).

#### 1.1.1 Finding closest centroids

In the “cluster assignment” phase of the K-means algorithm, the algorithm assigns every training example $x^{(i)}$ to its closest centroid, given the current positions of centroids. Specifically, for every example $i$ we set $c^{(i)} = j$ where $j = \min_{j} ||x^{(i)} - \mu_{j}||^{2}$. This step is implemented in the `findClosestCentroids` function below.

```python
#findClosestCentroids computes the centroid memberships for every example
def findClosestCentroids(X, centroids):
    # Initialize variables
    K = centroids.shape[0]
    num_examples = X.shape[0]
    idx = np.zeros(num_examples)
    
    # Run through all data points and all centroids
    for i in range(num_examples):
        min_dist = 1e6
        cluster = 0
        for j in range(K):
            dist = np.linalg.norm(X[i, :] - centroids[j, :]) ** 2
            if dist < min_dist:
                min_dist = dist
                cluster = j + 1
        idx[i] = cluster
        
    return idx
```

```python
# TESTS
K = 3
initial_centroids = np.array([[3, 3], [6, 2], [8, 5]])
idx = findClosestCentroids(X, initial_centroids)
print("Closest centroids for the first 3 examples: \n", idx[0:3])
print("Expected output: \n", np.array([1, 3, 2]))
```

#### 1.1.2 Computing centroid means

Given assignments of every point to a centroid, the second phase of the algorithm recomputes, for each centroid, the mean of the points that were assigned to it. Specifically, for every centroid $k$ we set $\mu_{k} = \frac{1}{|C_{k}|} \sum_{i \in C_{k}} x^{(i)}$, where $C_{k}$ is the set of examples that are assigned to centroid k. This step is implemented in the function `computeCentroids` below.

```python
#computeCentroids returns the new centroids by computing the means of the data points assigned to each centroid
def computeCentroids(X, idx):
    return np.array([np.mean(X[idx == i], axis = 0) for i in np.unique(idx)]) 
```

```python
# TESTS
centroids = computeCentroids(X, idx)
print("Centroids computed after initial finding of closest centroids: \n", centroids)
print("The centroids should be: \n", np.array([[2.428301, 3.157924], [5.813503, 2.633656], [7.119387, 3.616684]]))
```

### 1.2 Random Initialization

The initial assignments of centroids for the example dataset in the previous example were arbitrary. In practice, however, a good strategy for initializing the centroids is to select random examples from the training set. This is implemented in the function `kMeansInitCentroids`, shown below.

```python
#kMeansInitCentroids initializes K centroids that are to be used in K-Means on the dataset X
def kMeansInitCentroids(X, K):
    centroids = X[np.random.choice(X.shape[0], K, replace = False), :]
    
    return centroids
```

### 1.3 K-means on Example Dataset

After completing the three functions we can now run the K-means algorithm on a toy 2D dataset to help us understand how K-means works. We run the algorithm with the help of `runkMeans`, implemented below.

```python
#runkMeans runs the K-Means algorithm on data matrix X, where each row of X is a single example
def runkMeans(X, max_iters, K, plot_progress):
    # Initialize values
    num_examples, num_dim = X.shape
    centroids = kMeansInitCentroids(X, K)
    previous_centroids = centroids
    idx = np.zeros(num_examples)
    
    # Run K-Means
    for i in range(max_iters):
        # For each example in X, assign it to the closest centroid
        idx = findClosestCentroids(X, centroids)

        # Output plot progress
        if plot_progress:
            sns.scatterplot(x = X[:, 0], y = X[:, 1], hue = idx.astype(int), palette = "tab10", legend = False)
            plt.scatter(centroids[:, 0], centroids[:, 1], marker = "x", s = 30, linewidths = 2, color = "black")
            for j in range(centroids.shape[0]):
                plt.plot([centroids[j, 0], previous_centroids[j, 0]], [centroids[j, 1], previous_centroids[j, 1]], 'k-')
            previous_centroids = centroids
            plt.title("K-means Algorithm after " + str(max_iters) + " Runs")

        # Given the memberships, compute new centroids
        centroids = computeCentroids(X, idx)
        
    return centroids, idx
```

```python
max_iters = 10
centroids, idx = runkMeans(X, max_iters, 3, True)
```

`runkMeans` produces a visualization that steps you through the progress of the algorithm at each iteration.

### 1.4 Image Compression with K-means

In this section, we will apply K-means to image compression (to the image above, in particular). 

In a straightforward 24-bit color representation of an image, each pixel is represented as three 8-bit unsigned integers (ranging from 0 to 255) that specify the red, green and blue intensity values. This encoding is often refered to as the RGB encoding. Our image contains thousands of colors, and in this section, we will reduce the number of colors to 16 colors.

By making this reduction, it is possible to represent (compress) the photo in an efficient way. Specifically, we only need to store the RGB values of the 16 selected colors, and for each pixel in the image we now need to only
store the index of the color at that location (where only 4 bits are necessary to represent 16 possibilities).

We will use the K-means algorithm to select the 16 colors that will be used to represent the compressed image. Concretely, we will treat every pixel in the original image as a data example and use the K-means algorithm to find the 16 colors that best group (cluster) the pixels in the 3-dimensional RGB space. Once we have computed the cluster centroids on the image, we will then use the 16 colors to replace the pixels in the original image.

We first read the image to our notebook.

```python
# Load an image of a bird
A = plt.imread("bird_small.png")
```

This creates a three-dimensional matrix A whose first two indices identify a pixel position and whose last index represents red, green, or blue. For example, `A[49, 32, 2]` gives the blue intensity of the pixel at row 50 and column 33.

The code below reshapes the image to create an $m \times 3$ matrix of pixel colors (where $m$ = 16384 = 128 $\times$ 128), and calls our K-means function on it.

```python
# Save image size
A_shape = A.shape

# Reshape the image into an N x 3 matrix where N = number of pixels.
# Each row will contain the Red, Green and Blue pixel values
# This gives us our dataset matrix X that we will use K-Means on.
A = A.reshape(A_shape[0] * A_shape[1], A_shape[2]);

# Run K-Means algorithm on this data
K = 16
max_iters = 10
centroids, idx = runkMeans(A, max_iters, K, False)
```

After finding the top $K$ = 16 colors to represent the image, you can now assign each pixel position to its closest centroid using the `findClosestCentroids` function. This allows you to represent the original image using the centroid assignments of each pixel. 

```python
# Find closest cluster members
idx = findClosestCentroids(A, centroids)

# Essentially, now we have represented the image X as in terms of the indices in idx. We can now recover 
# the image from the indices (idx) by mapping each pixel(specified by its index in idx) to the centroid value
X_recovered = centroids[idx.astype(int) - 1]

# Reshape the recovered image into proper dimensions
X_recovered = X_recovered.reshape(A_shape[0], A_shape[1], A_shape[2])
```

Notice that you have significantly reduced the number of bits that are required to describe the image. The original image required 24 bits for each one of the 128$\times$128 pixel locations, resulting in total size of 128$\times$128$\times$24 = 393,216 bits. The new representation requires some overhead storage in form of a dictionary of 16 colors, each of which require 24 bits, but the image itself then only requires 4 bits per pixel location. The final number of bits used is therefore (16$\times$24)$+($128$\times$128$\times$4) = 65,920 bits, which corresponds to compressing the original image by about a factor of 6.

```python
# Show original and recovered image side-by-side
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(plt.imread("bird_small.png"))
ax1.axis("off")
ax1.set_title("Original Image")
ax2.imshow(X_recovered)
ax2.axis("off")
ax2.set_title("Recovered Image")
```

Finally, we can view the effects of the compression in the figure above. Even though the resulting image retains most of the characteristics of the original, we also see some compression artifacts.

## 2. Principal Components Analysis

In this exercise, we will use principal component analysis (PCA) to perform dimensionality reduction.
We will first experiment with an example 2D dataset to get intuition on how PCA works, and then use it on a bigger
dataset of 5000 face image dataset.

### 2.1 Example Dataset

We will first start with a 2D dataset which has one direction of large variation and one of smaller variation.

```python
# Read .mat file
data = io.loadmat('ex7data1.mat')
X = data["X"]

# Visualize the dataset
plt.scatter(X[:, 0], X[:, 1], marker = "o", color = "white", linewidths = 0.5, edgecolor = "blue")
```

In the succeeding sections, we will visualize what happens when we use PCA to reduce the data from 2D to 1D. In practice, you might want to reduce data from 256 to 50 dimensions, say; but using lower dimensional data in this example allows us to visualize the algorithms better.

### 2.2 Implementing PCA

PCA consists of two computational steps: First, you compute the covariance matrix of the data. Then, you use SVD to compute the eigenvectors $U_{1}, U_{2}, \ldots , U_{n}$. These will correspond to the principal components of
variation in the data.

Before using PCA, it is important to first normalize the data by subtracting the mean value of each feature from the dataset, and scaling each dimension so that they are in the same range. We do this using the `featureNormalize` function.

```python
# featureNormalize returns a normalized version of X where the mean value of each feature is 0 and the standard 
# deviation is 1. This is often a good preprocessing step to do when working with learning algorithms.
def featureNormalize(X):    
    # Get mean and standard deviation normalize
    mu = np.mean(X, 0)[np.newaxis, :]
    sigma = np.std(X, 0)[np.newaxis, :]

    # Normalize
    X_norm = (X - mu) / sigma

    return X_norm, mu, sigma
```

After normalizing the data, we can now run PCA to compute the principal components. 
- First, we should compute the covariance matrix of the data, which is given by $\Sigma = \frac{1}{m}X^{T}X$
- After computing the covariance matrix, you can run SVD on it to compute the principal components.

We do these steps in the `pca` function below.

```python
#pca runs principal component analysis on the dataset X
def pca(X):
    # Compute for the covariance matrix
    cov = 1 / X.shape[0] * (np.transpose(X) @ X)
    
    # Perform SVD on the covariance matrix
    U, S, V = np.linalg.svd(cov)
    
    return U, S
```

We now run PCA on the example dataset and plot the corresponding principal components found.

```python
# Plot data
plt.scatter(X[:, 0], X[:, 1], marker = "o", color = "white", linewidths = 0.5, edgecolor = "blue")

# Normalize the data
X_norm, mu, sigma = featureNormalize(X)

# Run PCA
U, S = pca(X_norm)

# Draw the eigenvectors centered at mean of data. These lines show the 
# directions of maximum variations in the dataset.
pt1 = mu
pt2 = mu + 1.5 * S[0] * U[:, 0].T
pt3 = mu + 1.5 * S[1] * U[:, 1].T
plt.plot([pt1[:, 0], pt2[:, 0]], [pt1[:, 1], pt2[:, 1]], '-k')
plt.plot([pt1[:, 0], pt3[:, 0]], [pt1[:, 1], pt3[:, 1]], '-k')

print("Top eigenvector: \n")
print("U[:,0] = ", U[0, 0], U[1, 0], "\n")
print("You should expect to see: -0.707107 -0.707107")
```

### 2.3 Dimensionality Reduction using PCA

After computing the principal components, we can use them to reduce the feature dimension of your dataset by projecting each example onto a lower dimensional space (e.g., projecting the data from 2D to 1D). In this section, we will use the eigenvectors returned by PCA and project the example dataset into a 1-dimensional space.

In practice, if you were using a learning algorithm such as linear regression or perhaps neural networks, you could now use the projected data instead of the original data. By using the projected data, you can train your model faster as there are less dimensions in the input.

#### 2.3.1 Projecting the data onto the principal components

We now implement the function `projectData`. Given a dataset $X$, the principal components $U$, and the desired number of dimensions to reduce to $K$, the function should project each example in $X$ onto the top $K$ components in $U$.

```python
#projectData computes the reduced data representation when projecting only on to the top k eigenvectors
def projectData(X, U, K):
    U_red = U[:, :K]
    return X @ U_red
```

```python
#TESTS
K = 1
Z = projectData(X_norm, U, K)
print("Projection of the first example: \n", Z[0, 0])
print("This value should be about 1.481274")
```

#### 2.3.2 Reconstructing an approximation of the data

After projecting the data onto the lower dimensional space, you can approximately recover the data by projecting them back onto the original high-dimensional space. This is implemented in the function `recoverData`.

```python
#recoverData recovers an approximation of the original data when using the projected data
def recoverData(Z, U, K):
    U_red = U[:, :K]
    return Z @ U_red.T
```

```python
#TESTS
X_rec  = recoverData(Z, U, K)
X_rec
print("Approximation of the first example: \n", X_rec[0, 0], X_rec[0, 1])
print("This value should be about -1.047419 -1.047419")
```

#### 2.3.2 Visualizing the projections

With `projectData` and `recoverData`, we can now perform both the projection and approximate reconstruction to show how the projection affects the data.

```python
plt.scatter(X_norm[:, 0], X_norm[:, 1], marker = "o", color = "white", linewidths = 0.5, edgecolor = "blue", label = "Original")
plt.scatter(X_rec[:, 0], X_rec[:, 1], marker = "o", color = "white", linewidths = 0.5, edgecolor = "red", label = "Projected")
plt.legend()

for i in range(X_norm.shape[0]):
    plt.plot([X_norm[i, 0], X_rec[i, 0]], [X_norm[i, 1], X_rec[i, 1]], color = "black", linewidth = 1, linestyle = "--")
```

In the figure above, the original (normalized) data points are indicated with the blue circles, while the projected data points are indicated with the red circles. The projection effectively only retains the information
in the direction given by $U_{1}$.

### 2.4 Face Image Dataset



We will now run PCA on face images to see how it can be used in practice for dimension reduction. The dataset `ex7faces.mat` contains a dataset X of face images, each 32 $\times$ 32 in grayscale. Each row of $X$ corresponds to one face image (a row vector of length 1024).

We first load the data onto our notebook.

```python
# Load faces dataset
faces = io.loadmat('ex7faces.mat')
faces = faces["X"]

# Plot faces
fig, ax = plt.subplots(nrows = 10, ncols = 10, figsize = (8, 8))
for i in range(0, 100, 10):
    for j in range(10):
        ax[int(i/10),j].imshow(faces[i + j, :].reshape(32, 32, order = "F"), cmap = "gray")
        ax[int(i/10),j].axis("off")
```

#### 2.4.1 PCA on Faces

To run PCA on the face dataset, we first normalize the dataset by subtracting the mean of each feature from the data matrix $X$. After running PCA, we obtain the principal components of the dataset.

```python
# Normalize the faces dataset
faces_norm, mu, sigma = featureNormalize(faces)

# Run PCA
U, S = pca(faces_norm)

# Get first 36 principal components
U_red = U[:, :36].T
```

Notice that each principal component in $U$ (each row) is a vector of length $n$ (where for the face dataset,
$n$ = 1024). It turns out that we can visualize these principal components by reshaping each of them into a 32$\times$32 matrix that corresponds to the pixels in the original dataset, as we have done below.

```python
# Visualize first 36 principal components
fig, ax = plt.subplots(nrows = 6, ncols = 6, figsize = (8, 8))
for i in range(0, 36, 6):
    for j in range(6):
        ax[int(i/6),j].imshow(U_red[i + j, :].reshape(32,32, order = "F"), cmap = "gray")
        ax[int(i/6),j].axis("off")
```

We can now project the face dataset onto only the first 100 principal components. Concretely, each face image is now described by a 100-dimensional vector. To understand what is lost in the dimension reduction, you can recover the data using only the projected dataset. 

Both projection and approximate recovery steps are performed below.

```python
# Project images to the eigenspace using the top 100 eigenvectors 
K = 100
Z = projectData(faces_norm, U, K)

# Project images to the eigenspace using the top 100 eigenvectors and visualize only using those 100 dimensions
K = 100
faces_rec  = recoverData(Z, U, K)
```

```python
# Visualize normalized face data
fig, ax = plt.subplots(nrows = 10, ncols = 10, figsize = (8, 8))
for i in range(0, 100, 10):
    for j in range(10):
        ax[int(i/10),j].imshow(faces_norm[i + j, :].T.reshape(32,32, order = "F"), cmap = "gray")
        ax[int(i/10),j].axis("off")
```

```python
# Visualize reconstructed face data
fig, ax = plt.subplots(nrows = 10, ncols = 10, figsize = (8, 8))
for i in range(0, 100, 10):
    for j in range(10):
        ax[int(i/10),j].imshow(faces_rec[i + j, :].T.reshape(32,32, order = "F"), cmap = "gray")
        ax[int(i/10),j].axis("off")
```

The original (normalized) and projected face images are displayed above. From the reconstruction, we can observe that the general structure and appearance of the face are kept while the fine details are lost. This is a remarkable reduction (more than 10 times) in the dataset size that can help speed up learning algorithms significantly. For example, if you were training a neural network to perform person recognition (gven a face image, predict the identitfy of the person), you can use the dimension reduced input of only a 100 dimensions instead of the original pixels.
